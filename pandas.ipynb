```python
#pandas: Series is a one dimensional array capable of holding any data type. 
#DataFrame is 2-D array capable of holding nay data type.
```

```python
#loc: Get rows(or columns) with particular labels(names) from the index
#iloc: gets rows or columns with particular positions in the index and it takes the integers.
```

```python
sports={"Football":"Spain",
        "Cricket": "India",
        "Athletes": "Africa",
        "Basketball": "USA"}
```

```python
sports
```

```python
import pandas as pd
```

```python
sports_series=pd.Series(sports)
```

```python
sports_series
```

```python
sports_series.loc["Cricket"]
```

```python
sports_series.iloc[3]
```
